# Import

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re
```

# Getting matches

## Big events

```python
match_page = []
event_types = ['BigEvents','Majors']
for event_type in event_types:
    for offset in range(0,951,50):
        url_matches = f'https://www.hltv.org/stats/matches?startDate=2019-01-01&endDate=2019-12-31&matchType={event_type}&offset={offset}'
        print(url_matches)
        html_matches = requests.get(url_matches)
        print(html_matches)
        tabela = pd.read_html(html_matches.content)[0]
        if pd.read_html(html_matches.content)[0].shape[0] == 0:
            print(offset+1)
        else:
            tabela['event_type'] = event_type
            tabela['event_tier'] = 'S'
            match_page.append(tabela)
matches = pd.concat(match_page)

```

```python
matches.reset_index(drop=True,inplace= True)
matches.to_csv('data/matches.csv',sep = ';')
```

# Players

## Majors

```python
url_players = 'https://www.hltv.org/stats/players?startDate=2019-01-01&endDate=2019-12-31&matchType=BigEvents&minMapCount=0'
html_players = requests.get(url_players)
soup_players = BeautifulSoup(html_players.content)
players_id = soup_players.select('body > div.bgPadding > div.widthControl > div.colCon > div.contentCol > div.stats-section > table > tbody > tr> td.playerCol > a')
player = [re.findall('/([0-9]*/.*)\?',str(i)) for i in players_id]
tabelas_players = []
for id in player:
    time.sleep(5)
    url_players = f'https://www.hltv.org/stats/players/matches/{id[0]}?startDate=2019-01-01&endDate=2019-12-31&matchType=Majors'
    print(url_players)
    html_players =(requests.get(url_players))
    print(html_players)
    tabela = pd.read_html(html_players.content)[0]
    tabela['player'] = id[0]
    tabela['event_type'] = 'major'
    tabela['event_tier'] = 'S'
    tabelas_players.append(tabela)
players_stats_majors = pd.concat(tabelas_players)
```
