# Pycaret modeling 

```python
import pandas as pd
from pycaret.regression import *
```

```python
players_stats = pd.read_csv('data/players_stats.csv' ,sep=';')
```

```python
players_stats
```

```python
#players_stats.drop(['date','opponent','player','player_team'],axis = 1,inplace = True)
```

```python
players_stats = players_stats.astype({'id':str})
```

```python
players_stats.info()
```

# General model

```python
model = setup(data = players_stats.drop(['id','date','player','opponent'],axis=1),target='rating2.0',session_id=42,normalize=True)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
save_model(best,'rating_pred')
```
